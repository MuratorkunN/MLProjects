```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df_mgm = pd.read_csv('data/mgm-bursa.csv')
df_open = pd.read_csv('data/open-meteo-7days.csv')
```

```python
print(df_mgm)
print(df_open)
```

```python
df_mgm['time'] = pd.to_datetime(df_mgm['time'])
df_open['time'] = pd.to_datetime(df_open['time'])
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df_mgm['time'], df_mgm['temperature_2m_min (°C)'],
         label='MGM Min', color='navy', linewidth=2, marker='o', markersize=6)
plt.plot(df_open['time'], df_open['temperature_2m_min (°C)'],
         label='Open-Meteo Min', color='cyan', linewidth=2, marker='o', markersize=6, linestyle='--')

plt.plot(df_mgm['time'], df_mgm['temperature_2m_max (°C)'],
         label='MGM Max', color='red', linewidth=2, marker='s', markersize=6)
plt.plot(df_open['time'], df_open['temperature_2m_max (°C)'],
         label='Open-Meteo Max', color='orange', linewidth=2, marker='s', markersize=6, linestyle='--')


plt.title('MGM vs Open-Meteo', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Temperature', fontsize=12)
plt.legend(loc='best')
plt.grid(True, linestyle=':', alpha=0.7)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())

plt.show()
```
